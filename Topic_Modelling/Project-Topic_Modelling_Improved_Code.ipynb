```python
# Opening and writing in a csv file in a loop:
def json_2_csv(dir_path):
    data_list=os.listdir(dir_path)
    c=0
    csv_file=open("csv_file.csv",'w',encoding='utf-8')
    csv_writer=csv.writer(csv_file)
    for data_fl in data_list:
        with open(os.path.join(data_path,data_fl)) as json_file:
            data = json.load(json_file)
        if c==0:
            csv_writer.writerow(data.keys())
            csv_writer.writerow(data.values())
            c+=1
        else:
            csv_writer.writerow(data.values())
    csv_file.close()
    return(os.path.abspath("csv_file.csv"))
```

```python
# Installing/ importing gensim and related libraries:
!pip install gensim
!pip install smart_open
import gensim
import gensim.corpora as corpora
```

```python
# Creating Dictionary:

#dataset = [d.split() for d in df_text['text']]
#id2word = corpora.Dictionary(dataset)
id2word = corpora.Dictionary([a.split() for a in df_text['text']])
```

```python
# Creating corpus:
dataset = [d.split() for d in df_text['text']]
```

```python
texts = dataset
```

```python
# Term document frequency:
corpus = [id2word.doc2bow(text) for text in texts]
```

# LDA Model Training:

Building a model keeping all the parameters to default where each topic is a combination of keywords and each keyword contributing certain weightage to the topic.

```python
# LDA model training:
from pprint import pprint

# Number of topics:
num_topics = 10

# Building LDA model:
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics)

# Printing the Keywords in the 10 topics:
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

The model having been trained, visualizing the topics for interpretability using pyLDAvis to interpret individual topics and better understand the relationship between the topics.
